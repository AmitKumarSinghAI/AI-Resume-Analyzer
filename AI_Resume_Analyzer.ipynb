```python
import google.generativeai as genai

genai.configure(api_key="PAST YOUR API KEY")
```

```python
import PyPDF2
```

```python
def read_resume(file_path):
    pdf = PyPDF2.PdfReader(file_path)
    text = ""

    for page in pdf.pages:
        text += page.extract_text()

    return text
```

```python
model = genai.GenerativeModel("gemini-2.5-flash")
```

```python
def resume_analyzer(resume_text):
    prompt = f"""
    You are an AI HR assistant.

    Analyze this resume and give:
    1. Skills
    2. Weak areas
    3. Suggestions to improve

    Resume:
    {resume_text}
    """

    response = model.generate_content(prompt)
    return response.text
```

```python
def job_match(resume_text, job_desc):
    prompt = f"""
    Compare resume with job description.

    Give:
    1. Match percentage (0-100)
    2. Missing skills
    3. Recommendation

    Resume:
    {resume_text}

    Job Description:
    {job_desc}
    """

    response = model.generate_content(prompt)
    return response.text
```

```python
def generate_questions(resume_text, job_desc):
    prompt = f"""
    Generate 10 interview questions based on:

    Resume:
    {resume_text}

    Job:
    {job_desc}
    """

    response = model.generate_content(prompt)
    return response.text
```

```python
def mock_interview(question):
    print("Question:", question)
    answer = input("Your Answer: ")

    prompt = f"""
    You are an interviewer.

    Question: {question}
    Candidate Answer: {answer}

    Give rating out of 10 and feedback.
    """

    response = model.generate_content(prompt)
    return response.text
```

```python
# Upload resume manually in Colab
resume_text = read_resume("/content/Amit_Kumar_singh_Data_Science_resume.pdf")

job_desc = input("JOb Description : ")

print("\n=== RESUME ANALYSIS ===")
print(resume_analyzer(resume_text))

print("\n=== JOB MATCH ===")
print(job_match(resume_text, job_desc))

print("\n=== INTERVIEW QUESTIONS ===")
questions = generate_questions(resume_text, job_desc)
print(questions)
```
